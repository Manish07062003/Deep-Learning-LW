```python
import pandas as pd
```

```python
dataset = pd.read_csv("Churn_Modelling.csv")
```

```python
dataset
```

```python
y = dataset["Exited"]
y
```

```python
dataset.columns
```

```python
x = dataset[['RowNumber', 'CustomerId', 'Surname', 'CreditScore', 'Geography',
       'Gender', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard',
       'IsActiveMember', 'EstimatedSalary']]
x
```

```python
geo = pd.get_dummies(x["Geography"],drop_first=True,dtype=int)
geo
```

```python
gender = pd.get_dummies(x["Gender"],drop_first=True,dtype=int)
gender
```

```python
NumOfProducts = pd.get_dummies(x["NumOfProducts"],drop_first=True,dtype=int)
NumOfProducts
```

```python
HasCrCard = pd.get_dummies(x["HasCrCard"],drop_first=True,dtype=int)
HasCrCard
```

```python
IsActiveMember = pd.get_dummies(x["IsActiveMember"],drop_first=True,dtype=int)
IsActiveMember
```

```python
x.drop(['Geography','Gender',"IsActiveMember",'HasCrCard','NumOfProducts','RowNumber','CustomerId','Surname'],axis=1,inplace=True)
```

```python
x
```

```python
new_x = pd.concat([x,geo,gender,NumOfProducts,IsActiveMember,HasCrCard],axis=1)
new_x
```

```python
from keras.layers import Dense
```

```python
from keras.models import Sequential
```

```python
brain = Sequential()
```

```python
# layer 1
# here the problem statement is a classification problem, so we should set the activation function as sigmoid,
# but here in neural network for intermediate function the activation function should be relu and the final output
# layer function should be sigmoid as the final layer should give 0 or 1
brain.add(
    Dense(
        units=6,
        input_dim=13,
        activation='relu',
        bias_initializer='zeros',
        kernel_initializer='zeros',
    )
)
```

```python
# here total params is 84 because every features is given to each neuron(i.e, machine learning model) so there are
# 6 neurons and 13 features + 1 bias so 6*14=84 params.Each neurons gives different weights for each feature (x).
brain.summary()
```

```python
# layer 2
# for second layer there is no need to specify the input shape, the neurons will take care of it.
brain.add(
    Dense(
        units=7,
        activation='relu',
        bias_initializer='zeros',
        kernel_initializer='zeros',
    )
)
```

```python
# here for second layer the total params are 49 because there are 6 neurons in first layers, so six neuron combinely 
# gives their trained information to each neuron in layer 2 (7 neurons) + 1 bias , so (6+1)*7 = 49
brain.summary()
```

```python
# layer 3
brain.add(
    Dense(
        units=5,
        activation='relu',
        bias_initializer='zeros',
        kernel_initializer='zeros',
    )
)
```

```python
brain.summary()
```

```python
# adding the final layer
brain.add(
    Dense(
        units=1,
        activation='sigmoid',
        bias_initializer='zeros',
        kernel_initializer='zeros',
    )
)
```

```python
brain.get_config()
```

```python
brain.compile(loss="binary_crossentropy")
```

```python
brain.fit(new_x,y,epochs=10)
```

```python
brain.get_weights()
```

```python
brain.predict(new_x)
```
