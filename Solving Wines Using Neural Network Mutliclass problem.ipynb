```python
import pandas as pd
```

```python
dataset = pd.read_csv("wine.csv")
dataset
```

## The above dataset is a multi class problem as y (Wine) has 3 possible classes (1,2,3).

```python
y = dataset["Wine"]
y
```

## Based on the composition of every feature the model predicts which class of wine will be the output

```python
import seaborn as sns
```

## In multiclass problem, the prediction will not be done as single value , there will be a probability of every class, so whose probablity is more we can choose that class.

## In multiclass, y will be multiple not a single value. (since it is a categorical variable)

```python
y = pd.get_dummies(dataset["Wine"],dtype=int)
y
```

```python
dataset.columns
```

```python
x = dataset[['Alcohol', 'Malic.acid', 'Ash', 'Acl', 'Mg', 'Phenols',
       'Flavanoids', 'Nonflavanoid.phenols', 'Proanth', 'Color.int', 'Hue',
       'OD', 'Proline']]
x
```

## There are no categorical variable in x

```python
from keras.layers import Dense
```

```python
from keras.models import Sequential
```

```python
brain = Sequential()
```

```python
brain.get_config()
```

```python
# adding layer 1
brain.add(
    Dense(
        units=5,
        input_dim=13,
        activation = 'relu',
        bias_initializer='zeros',
        kernel_initializer='zeros'
    )
)
```

```python
brain.summary()
```

```python
# adding layer 2
brain.add(
    Dense(
        units=5,
        activation = 'relu',
        bias_initializer='zeros',
        kernel_initializer='zeros'
    )
)
```

```python
brain.summary()
```

```python
# adding output layer 3
# since there are 3 y's so we want 3 outputs thats why we are using 3 neurons in the output layer
brain.add(
    Dense(
        units=3,
        activation = 'softmax',
        bias_initializer='zeros',
        kernel_initializer='zeros'
    )
)
```

```python
brain.summary()
```

```python
brain.compile(loss='categorical_crossentropy')
```

```python
brain.fit(x,y,epochs=20)
```

```python
brain.get_weights()
```
